# Sinteze dos Benefícios

# Sinteze dos benefícios em classificações gerais encontradas

```python
import pandas as pd
```

```python
dados_beneficios_gerais = pd.read_csv('Sinteze - analise_beneficios_gerais.csv')
df_beneficios_gerais = pd.DataFrame(dados_beneficios_gerais)
```

```python
df_beneficios_gerais
```

## Gráfico de pizza da distribuição dos benefícios gerais encontrados nos estudos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Configuração de cores e estilo
cores = plt.get_cmap('Paired')(np.linspace(0, 1, len(df_beneficios_gerais)))

plt.figure(figsize=(12, 7))

#Criando o gráfico
fatias, textos, autotextos = plt.pie(
    df_beneficios_gerais['Total'],
    labels=None,
    autopct='%1.1f%%',
    startangle=140,
    colors=cores,
    pctdistance=0.85,
    explode=[0.03] * len(df_beneficios_gerais)
)

#Ajustando a fonte das porcentagens para ficarem legíveis
plt.setp(autotextos, size=9, weight="bold", color="black")

#Adicionando legenda lateral para não poluir o gráfico
plt.legend(
    fatias,
    df_beneficios_gerais['Benefícios'],
    title="Benefícios",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title('Distribuição de Benefícios', fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```

# Sinteze dos benefícios por etapa do Ciclo de Desenvolvimento de Software

```python
dados_beneficios_SDLC = pd.read_csv('Sinteze - analise_por_etapa_do_SDLC.csv')
df_beneficios_SDLC = pd.DataFrame(dados_beneficios_SDLC)
```

```python
df_beneficios_SDLC
```

## Gráfico de pizza para a distribuição dos benefícios por etapa do Ciclo de Desenvolvimento de Software

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Configurando cores
cores = plt.get_cmap('Paired')(np.linspace(0, 1, len(df_beneficios_SDLC)))

plt.figure(figsize=(12, 7))

#Criando o gráfico
fatias, textos, autotextos = plt.pie(
    df_beneficios_SDLC['Numero de estudos'],
    labels=None,
    autopct='%1.1f%%',
    startangle=140,
    colors=cores,
    pctdistance=0.85,
    explode=[0.03] * len(df_beneficios_SDLC)
)

#Ajustando a fonte das porcentagens
plt.setp(autotextos, size=9, weight="bold", color="black")

#Adicionando legenda lateral
plt.legend(
    fatias,
    df_beneficios_SDLC['Classificações de Beneficios'],
    title="Benefícios",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title('Distribuição de Benefícios por etapa do SDLC', fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```

## Gráfico de barras para as classificações dos benefícios por etapa do Ciclo de Desenvolvimento de Software

```python
import matplotlib.pyplot as plt
import numpy as np

categorias = df_beneficios_SDLC['Classificações de Beneficios']
valores = df_beneficios_SDLC['Numero de estudos']

#Configurando cores
cores = plt.get_cmap('Paired')(np.linspace(0, 1, len(df_beneficios_SDLC)))

#Espaçamento entre barras
posicoes = np.arange(len(categorias)) * 1.2

#Criando o gráfico
plt.figure(figsize=(8, 5))
plt.bar(posicoes, valores, width=0.8, color=cores)

#Personalização
plt.title('Distribuição de Benefícios por etapa do SDLC')
plt.xlabel('Benefícios')
plt.ylabel('Numero de estudos')
plt.xticks(posicoes, categorias, rotation=45, ha='right')

#Mostrando
plt.tight_layout()
plt.show()
```
